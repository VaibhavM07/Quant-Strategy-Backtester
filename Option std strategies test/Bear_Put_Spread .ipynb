```python
#Importing all the libs
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import datetime
import glob
pd.options.mode.chained_assignment = None 
```

```python
folder = glob.glob("C:\\Users\\vaibh\\Desktop\\file(01)")
folder
```

```python
for file in folder:
    files = glob.glob(file + "/*.csv")
files
```

```python
tradelog = pd.DataFrame(columns = ['Ticker', 'Entry Time', 'Entry Price', 'Stop Loss Exit', 'Exit Time', 'Exit Price'])
tradelog
```

```python
xx = pd.read_csv(files[0])
xx.insert(0,'Timestamp',xx['Date'] + " " + xx['Time'])
xx['Timestamp'] = pd.to_datetime(xx['Timestamp'] , format = '%d/%m/%Y %H:%M:%S')
xx.drop('Time',axis = 1 , inplace = True)
xx.drop('Date',axis = 1 , inplace = True)
xx
```

```python
ticker = 'BANKNIFTY'
data = xx[xx['Ticker'].apply(lambda a: a[0:len(ticker)]==ticker)]
data
```

```python
puts = data[data['Ticker'].apply(lambda a : a.strip()[-2:] == 'PE')]
puts
```

```python
futures = data[data['Ticker'].apply(lambda a : a.strip()[-2:] == '-I')]
futures
```

```python
expiry = puts[puts['Volume'] == puts['Volume'].max()]['Ticker'].iloc[0].strip()[:len(ticker)+5]
expiry
```

```python
current_expiry = puts[puts['Ticker'].apply(lambda a: a.strip()[:len(ticker)+5] == expiry)]
current_expiry
```

```python
ATM = int(futures['Open'].iloc[0].round(-2))
ATM
```

```python
ITM = ATM + 200
ITM
```

```python
OTM = ATM - 200
OTM
```

```python
ITM_strike = str(ITM) + 'PE'
ITM_strike
```

```python
OTM_strike = str(OTM) + 'PE'
OTM_strike
```

```python
ITM_puts = current_expiry[current_expiry['Ticker'].apply(lambda a : a.strip()[-7:] == ITM_strike)]
ITM_puts.columns = ["Timestamp", "PE_Ticker_ITM", "PE_Open_ITM", "PE_High_ITM", "PE_Low_ITM", "PE_Close_ITM", "PE_Volume_ITM", "PE_Open_Interest_ITM"]
ITM_puts
```

```python
OTM_puts = current_expiry[current_expiry['Ticker'].apply(lambda a : a.strip()[-7:] == OTM_strike)]
OTM_puts.columns = ["Timestamp", "PE_Ticker_OTM", "PE_Open_OTM", "PE_High_OTM", "PE_Low_OTM", "PE_Close_OTM", "PE_Volume_OTM", "PE_Open_Interest_OTM"]
OTM_puts
```

```python
trade_data = pd.merge(ITM_puts,OTM_puts)
trade_data
```

```python
Entry_time = "09:29:59"
trade_data = trade_data[trade_data['Timestamp'] >= pd.to_datetime(str(trade_data['Timestamp'].iloc[0].date())+ " "+ Entry_time)]
trade_data = trade_data[trade_data['Timestamp'] < pd.to_datetime(str(trade_data['Timestamp'].iloc[0].date())+ " "+ "15:14:59")]
trade_data.reset_index(drop=True, inplace=True)
trade_data
```

```python
#stoploss
PE_ticker_ITM = trade_data['PE_Ticker_ITM'].iloc[0]
PE_ticker_OTM = trade_data['PE_Ticker_OTM'].iloc[0]
PE_ITM_Entry_time = PE_OTM_Entry_time = trade_data['Timestamp'].iloc[0]
PE_ITM_Entry_price = trade_data['PE_Open_ITM'].iloc[0]
PE_OTM_Entry_price = trade_data['PE_Open_OTM'].iloc[0]

stop_loss_price_ITM = (PE_ITM_Entry_price)*((20+100)/100)
stop_loss_price_OTM = (PE_OTM_Entry_price)*((100-20)/100)
stop_loss = trade_data[(trade_data['PE_Open_ITM'] > stop_loss_price_ITM) | (trade_data['PE_Open_OTM'] < stop_loss_price_OTM)]
display(stop_loss)
```

```python
if stop_loss.empty:
    stop_loss_exit = 0
    PE_ITM_exit_time = PE_OTM_exit_time = trade_data['Timestamp'].iloc[-1]
    PE_ITM_exit_price = trade_data['PE_Open_ITM'].iloc[-1]
    PE_OTM_exit_price = trade_data['PE_Open_OTM'].iloc[-1]
else:
    stop_loss_exit = 1
    PE_ITM_exit_time = PE_OTM_exit_time = stop_loss['Timestamp'].iloc[0]
    PE_ITM_exit_price = stop_loss['PE_Open_ITM'].iloc[0]
    PE_OTM_exit_price = stop_loss['PE_Open_OTM'].iloc[0]
```

```python
#Entry to tradelog
tradelog.loc[len(tradelog)] = [PE_ticker_ITM, PE_ITM_Entry_time, PE_ITM_Entry_price, stop_loss_exit, PE_ITM_exit_time, PE_ITM_exit_price]
tradelog.loc[len(tradelog)] = [PE_ticker_OTM, PE_OTM_Entry_time, PE_OTM_Entry_price, stop_loss_exit, PE_OTM_exit_time, PE_OTM_exit_price]
tradelog
```

```python
tradelog['Buy'] = 0
tradelog['Sell'] = 0
for i in range(len(tradelog)):
    
    if tradelog['Ticker'].iloc[i].strip()[-7:] == OTM_strike:
        tradelog['Buy'][i] = 1
        tradelog['Sell'][i] = 0
    elif tradelog['Ticker'].iloc[i].strip()[-7:] == ITM_strike:
        tradelog['Sell'][i] = 1
        tradelog['Buy'][i] = 0
tradelog
```

```python
#PNL
tradelog['PnL'] = 0
for i in range(len(tradelog)):
    if tradelog['Buy'][i] == 1 and tradelog['Sell'][i] == 0 :
        tradelog['PnL'][i] = 25*(tradelog['Exit Price'][i] - tradelog['Entry Price'][i])
    elif tradelog['Buy'][i] == 0 and tradelog['Sell'][i] == 1 :
        tradelog['PnL'][i] = 25*(tradelog['Entry Price'][i] - tradelog['Exit Price'][i])
    else:
        pass
tradelog['Cumulative_PnL'] = tradelog['PnL'].cumsum()
tradelog
```
