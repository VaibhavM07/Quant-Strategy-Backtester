```python
df.insert(0,'Timestamp',df['Date'] + " " + df['Time'])
df['Timestamp'] = pd.to_datetime(df['Timestamp'] , format = '%d/%m/%Y %H:%M:%S')
df.drop('Date' , axis = 1 , inplace = True)
df.drop('Time' , axis = 1 , inplace = True)
df
```

```python
ticker = 'BANKNIFTY-I'
df = df[df['Ticker'].apply(lambda a : a[0:len(ticker)] == ticker)]
df.drop("Open Interest",axis = 1 , inplace = True)
df
```

```python
ohlc = {
        'Ticker':'first',
        'Open': 'first',
        'High': 'max',
        'Low':'min',
        'Close':'last',
        'Volume':'sum'
        }
    
for key in df:
            
    df.set_index('Timestamp', inplace=True)
    df = df.resample('5min').apply(ohlc)
    df.dropna(inplace=True)
    df.reset_index(inplace=True)
print(df)
```

```python

df["MA"] = df["Close"].ewm(span=20,min_periods = 20).mean()
df["BB_up"] = df["MA"]+ 2*df["Close"].rolling(20).std(ddof=0)
df["BB_down"] = df["MA"]- 2*df["Close"].rolling(20).std(ddof=0)
df["BB_width"] = df["BB_up"]-df["BB_down"]
#signal only when K & D is above 80 or beLow 20 - buy when above 80 and K>D - opposite for sell
df['14-High'] = df['High'].rolling(14).max()
df['14-Low'] = df['Low'].rolling(14).min()
df['%K'] = (df['Close'] - df['14-Low'])*100/(df['14-High'] - df['14-Low'])
df['%D'] = df['%K'].rolling(3).mean()
df["5EMA"] = df["Close"].ewm(span=5,min_periods=5).mean()
#df.dropna(inplace = True)
df.reset_index(inplace = True)
with pd.option_context('display.max_rows', None, 'display.max_columns', None):  
    display(df)
#df.plot(x="Timestamp", y=['Close','BB_up' , 'BB_down','5EMA'])
#df.plot(x="Timestamp", y=['%K',"%D"])
```

```python
tradelog = pd.DataFrame(columns= ["Entry_Time","Signal","Exit_Time","Target_hit","Stoploss_hit"])
k = 0
for i in range(1,len(df)):
    if (df["Close"][i-1]+(df["Close"][i-1]*0.00025))>=df["BB_up"][i] and (df["Low"][i-1]-5)>df["5EMA"][i]:
        tradelog.loc[k,'Signal'] = "ATM_PUT_BUY"
        tradelog.loc[k,'Entry_Time'] = df['Timestamp'][i]
        for j in range(i,10):
            if df["Close"][j]<= df["5EMA"][j]:
                tradelog.loc[k,'Exit_Time'] = df['Timestamp'][j]
                tradelog.loc[k,'Target_hit'] = 1
                tradelog.loc[k,'Stoploss_hit'] = 0
            elif df["Close"][j] >= df["High"][j-1]+(df["High"][j-1]-df["Low"][j-1]):
                tradelog.loc[k,'Exit_Time'] = df['Timestamp'][j]
                tradelog.loc[k,'Stoploss_hit'] = 1
                tradelog.loc[k,'Target_hit'] = 0

        
    elif (df["Close"][i-1]-(df["Close"][i-1]*0.00025))<df["BB_down"][i] and (df["High"][i-1]+5)<df["5EMA"][i]:
        tradelog.loc[k,'Signal'] = "ATM_CALL_BUY"
        tradelog.loc[k,'Entry_Time'] = df['Timestamp'][i]
        for j in range(i,10):
            if df["Close"][j]>= df["5EMA"][j]:
                tradelog.loc[k,'Exit_Time'] = df['Timestamp'][j]
                tradelog.loc[k,'Target_hit'] = 1
                tradelog.loc[k,'Stoploss_hit'] = 0
            elif df["Close"][j] <= df["High"][j-1]-(df["High"][j-1]-df["Low"][j-1]):
                tradelog.loc[k,'Exit_Time'] = df['Timestamp'][j]
                tradelog.loc[k,'Stoploss_hit'] = 1
                tradelog.loc[k,'Target_hit'] = 0
    k+=1
#tradelog .reset_index(drop = True,inplace = True)
#with pd.option_context('display.max_rows', None, 'display.max_columns', None):  
tradelog


```
