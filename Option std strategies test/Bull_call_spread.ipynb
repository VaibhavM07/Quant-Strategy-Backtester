```python
tradelog = pd.DataFrame(columns = ['Ticker', 'Entry Time', 'Entry Price', 'Stop Loss Exit', 'Exit Time', 'Exit Price'])
tradelog
```

```python
xx = pd.read_csv(files[1])
xx.insert(0,'Timestamp',xx['Date'] + " " + xx["Time"])
xx['Timestamp'] = pd.to_datetime(xx['Timestamp'] , format = '%d/%m/%Y %H:%M:%S')
xx.drop('Date',axis = 1 , inplace = True)
xx.drop('Time' , axis = 1 , inplace = True)
xx
```

```python
#Because ticker includes int and string value aslo we are converting whole ticker into string
xx['Ticker'] = xx['Ticker'].apply(str)
```

```python
#getting only the banknifty data
ticker = 'BANKNIFTY'    
data = xx[xx['Ticker'].apply(lambda a : a[0:len(ticker)]==ticker)]
data
```

```python
#getting all calls, puts  and futures data
calls  = data[data['Ticker'].apply(lambda a : a.strip()[-2:] == 'CE')]
futures = data[data['Ticker'].apply(lambda a: a.strip()[-2:]=="-I")]
futures
```

```python
#Getting current expiry for calls
calls_prefix = calls[calls['Volume'] == calls['Volume'].max()]['Ticker'].iloc[0][:len(ticker)+5]
calls_prefix
calls_current_expiry = calls[calls['Ticker'].apply(lambda a : a[0:len(ticker)+5]==calls_prefix)]
calls_current_expiry
```

```python
strike_ATM = int(futures['Open'].iloc[0].round(-2))
strike_ATM
```

```python
strike_OTM = strike_ATM + 200
strike_OTM
```

```python
#adding the strikes
OTM_strike = str(strike_OTM) + "CE"
ATM_strike = str(strike_ATM) + "CE"
```

```python
ATM_calls = calls_current_expiry[calls_current_expiry['Ticker'].apply(lambda a:a.strip()[-7:]==ATM_strike)]
#ATM_calls.columns= ["Timestamp", "CE_Ticker_ATM", "CE_Open_ATM", "CE_High_ATM", "CE_Low_ATM", "CE_Close_ATM", "CE_Volume_ATM", "CE_Open_Interest_ATM"]
ATM_calls
```

```python
OTM_calls = calls_current_expiry[calls_current_expiry['Ticker'].apply(lambda a:a.strip()[-7:]==OTM_strike)]
OTM_calls.columns= ["Timestamp", "CE_Ticker_OTM", "CE_Open_OTM", "CE_High_OTM", "CE_Low_OTM", "CE_Close_OTM", "CE_Volume_OTM", "CE_Open_Interest_OTM"]
OTM_calls
```

```python
trade_data = pd.merge(ATM_calls,OTM_calls, on = 'Timestamp')
trade_data
```

```python
#SHORTLISTING DATA BASED ON ENTRY AND EXIT TIME ONLY
Entry_time = "09:29:59"
trade_data = trade_data[trade_data['Timestamp'] >= pd.to_datetime(str(trade_data["Timestamp"].iloc[0].date())+" "+Entry_time)]
trade_data = trade_data[trade_data["Timestamp"] < pd.to_datetime(str(trade_data["Timestamp"].iloc[0].date())+" "+"15:15:00")]
trade_data.reset_index(drop=True, inplace=True)
trade_data
```

```python
#STOPLOSS LOGIC
CE_ticker_ATM = trade_data['CE_Ticker_ATM'].iloc[0]
CE_ticker_OTM = trade_data['CE_Ticker_OTM'].iloc[0]
CE_ATM_Entry_time = CE_OTM_Entry_time = trade_data['Timestamp'].iloc[0]
CE_ATM_Entry_price = trade_data['CE_Open_ATM'].iloc[0]
CE_OTM_Entry_price = trade_data['CE_Open_OTM'].iloc[0]

stop_loss_price_ATM = (CE_ATM_Entry_price)*((100-20)/100)
stop_loss_price_OTM = (CE_OTM_Entry_price)*((20+100)/100)
stop_loss = trade_data[(trade_data['CE_Open_ATM'] < stop_loss_price_ATM) | (trade_data['CE_Open_OTM'] > stop_loss_price_OTM)]
display(stop_loss)
```

```python
if stop_loss.empty:
    stop_loss_exit = 0
    CE_ATM_exit_time = CE_OTM_exit_time = trade_data['Timestamp'].iloc[-1]
    CE_ATM_exit_price = trade_data['CE_Open_ATM'].iloc[-1]
    CE_OTM_exit_price = trade_data['CE_Open_OTM'].iloc[-1]
else:
    stop_loss_exit = 1
    CE_ATM_exit_time = CE_OTM_exit_time = stop_loss['Timestamp'].iloc[0]
    CE_ATM_exit_price = stop_loss['CE_Open_ATM'].iloc[0]
    CE_OTM_exit_price = stop_loss['CE_Open_OTM'].iloc[0]
```

```python
#Entry to tradelog
tradelog.loc[len(tradelog)] = [CE_ticker_ATM, CE_ATM_Entry_time, CE_ATM_Entry_price, stop_loss_exit, CE_ATM_exit_time, CE_ATM_exit_price]
tradelog.loc[len(tradelog)] = [CE_ticker_OTM, CE_OTM_Entry_time, CE_OTM_Entry_price, stop_loss_exit, CE_OTM_exit_time, CE_OTM_exit_price]
```

```python
tradelog
```

```python
tradelog['Buy'] = 0
tradelog['Sell'] = 0
for i in range(len(tradelog)):
    
    if tradelog['Ticker'].iloc[i].strip()[-7:] == ATM_strike:
        tradelog['Buy'][i] = 1
        tradelog['Sell'][i] = 0
    elif tradelog['Ticker'].iloc[i].strip()[-7:] == OTM_strike:
        tradelog['Sell'][i] = 1
        tradelog['Buy'][i] = 0
tradelog
```

```python
#PNL
tradelog['PnL'] = 0
for i in range(len(tradelog)):
    if tradelog['Buy'][i] == 1 and tradelog['Sell'][i] == 0 :
        tradelog['PnL'][i] = 25*(tradelog['Exit Price'][i] - tradelog['Entry Price'][i])
    elif tradelog['Buy'][i] == 0 and tradelog['Sell'][i] == 1 :
        tradelog['PnL'][i] = 25*(tradelog['Entry Price'][i] - tradelog['Exit Price'][i])
    else:
        pass
tradelog['Cumulative_PnL'] = tradelog['PnL'].cumsum()
tradelog
```
