```python
#Importing all the libs
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import datetime
import glob
pd.options.mode.chained_assignment = None  
```

```python
for file in folder:
    files = glob.glob(file + '/*.csv')
files
```

```python
tradelog = pd.DataFrame(columns = ['Ticker', 'Entry Time', 'Entry Price', 'Stop Loss Exit', 'Exit Time', 'Exit Price'])
tradelog
```

```python
xx = pd.read_csv(files[0])
xx.insert(0,'Timestamp',xx['Date'] + " " + xx['Time'])
xx['Timestamp'] = pd.to_datetime(xx['Timestamp'] , format = '%d/%m/%Y %H:%M:%S')
xx.drop('Date',axis = 1 , inplace = True)
xx.drop('Time' , axis = 1 , inplace = True)
xx
```

```python
ticker = 'BANKNIFTY'
xx['Ticker'] = xx['Ticker'].apply(str)
data = xx[xx['Ticker'].apply(lambda a : a[0:len(ticker)] == ticker)]
data
```

```python
calls = data[data['Ticker'].apply(lambda a : a.strip()[-2:] == 'CE')]
futures = data[data['Ticker'].apply(lambda a: a.strip()[-2:] == '-I')]
```

```python
expiry = calls[calls['Volume'] == calls['Volume'].max()]['Ticker'].iloc[0][:len(ticker)+5]
expiry
```

```python
calls = calls[calls['Ticker'].apply(lambda a : a[0:len(ticker)+5] == expiry)]
calls
```

```python
strike = int(futures['Open'].iloc[0].round(-2))
ATM_strike = str(strike)+'CE'
ATM_strike
```

```python
ITM_strike = strike - 200
ITM_strike = str(ITM_strike) + 'CE'
ITM_strike
```

```python
OTM_strike = strike + 200
OTM_strike = str(OTM_strike) + 'CE'
OTM_strike
```

```python
ITM_calls = calls[calls['Ticker'].apply(lambda a : a.strip()[-7:] == ITM_strike)]
ITM_calls.columns = ["Timestamp", "CE_Ticker_ITM", "CE_Open_ITM", "CE_High_ITM", "CE_Low_ITM", "CE_Close_ITM", "CE_Volume_ITM", "CE_Open_Interest_ITM"]
ITM_calls
```

```python
OTM_calls = calls[calls['Ticker'].apply(lambda a : a.strip()[-7:] == OTM_strike)]
OTM_calls.columns = ["Timestamp", "CE_Ticker_OTM", "CE_Open_OTM", "CE_High_OTM", "CE_Low_OTM", "CE_Close_OTM", "CE_Volume_OTM", "CE_Open_Interest_OTM"]
OTM_calls
```

```python
trade_data = pd.merge(OTM_calls,ITM_calls)
trade_data
```

```python
Entry_time = "09:29:59"
trade_data = trade_data[trade_data['Timestamp'] >= pd.to_datetime(str(trade_data['Timestamp'].iloc[0].date())+ " "+ Entry_time)]
trade_data = trade_data[trade_data['Timestamp'] < pd.to_datetime(str(trade_data['Timestamp'].iloc[0].date())+ " "+ "15:14:59")]
trade_data.reset_index(drop=True, inplace=True)
trade_data
```

```python
#STOPLOSS LOGIC
CE_ticker_ITM = trade_data['CE_Ticker_ITM'].iloc[0]
CE_ticker_OTM = trade_data['CE_Ticker_OTM'].iloc[0]
CE_ITM_Entry_time = CE_OTM_Entry_time = trade_data['Timestamp'].iloc[0]
CE_ITM_Entry_price = trade_data['CE_Open_ITM'].iloc[0]
CE_OTM_Entry_price = trade_data['CE_Open_OTM'].iloc[0]

stop_loss_price_ITM = (CE_ITM_Entry_price)*((100+20)/100)
stop_loss_price_OTM = (CE_OTM_Entry_price)*((100-20)/100)
stop_loss = trade_data[(trade_data['CE_Open_ITM'] > stop_loss_price_ITM) | (trade_data['CE_Open_OTM'] < stop_loss_price_OTM)]
display(stop_loss)
```

```python
if stop_loss.empty:
    stop_loss_exit = 0
    CE_ITM_exit_time = CE_OTM_exit_time = trade_data['Timestamp'].iloc[-1]
    CE_ITM_exit_price = trade_data['CE_Open_ITM'].iloc[-1]
    CE_OTM_exit_price = trade_data['CE_Open_OTM'].iloc[-1]
else:
    stop_loss_exit = 1
    CE_ITM_exit_time = CE_OTM_exit_time = stop_loss['Timestamp'].iloc[0]
    CE_ITM_exit_price = stop_loss['CE_Open_ITM'].iloc[0]
    CE_OTM_exit_price = stop_loss['CE_Open_OTM'].iloc[0]
```

```python
#Entry to tradelog
tradelog.loc[len(tradelog)] = [CE_ticker_ITM, CE_ITM_Entry_time, CE_ITM_Entry_price, stop_loss_exit, CE_ITM_exit_time, CE_ITM_exit_price]
tradelog.loc[len(tradelog)] = [CE_ticker_OTM, CE_OTM_Entry_time, CE_OTM_Entry_price, stop_loss_exit, CE_OTM_exit_time, CE_OTM_exit_price]
tradelog
```

```python
tradelog['Buy'] = 0
tradelog['Sell'] = 0
for i in range(len(tradelog)):
    
    if tradelog['Ticker'].iloc[i].strip()[-7:] == OTM_strike:
        tradelog['Buy'][i] = 1
        tradelog['Sell'][i] = 0
    elif tradelog['Ticker'].iloc[i].strip()[-7:] == ITM_strike:
        tradelog['Sell'][i] = 1
        tradelog['Buy'][i] = 0
tradelog
```

```python
#PNL
tradelog['PnL'] = 0
for i in range(len(tradelog)):
    if tradelog['Buy'][i] == 1 and tradelog['Sell'][i] == 0 :
        tradelog['PnL'][i] = 25*(tradelog['Exit Price'][i] - tradelog['Entry Price'][i])
    elif tradelog['Buy'][i] == 0 and tradelog['Sell'][i] == 1 :
        tradelog['PnL'][i] = 25*(tradelog['Entry Price'][i] - tradelog['Exit Price'][i])
    else:
        pass
tradelog['Cumulative_PnL'] = tradelog['PnL'].cumsum()
tradelog
```
