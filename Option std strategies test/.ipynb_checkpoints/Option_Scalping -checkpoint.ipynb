```python
ATM_calls = calls[calls['Ticker'].apply(lambda a : a.strip()[-7:] == ATM_call_strike)]
ATM_calls
```

```python
Entry_time = "09:29:59"
ATM_calls = ATM_calls[ATM_calls['Timestamp'] >= pd.to_datetime(str(ATM_calls['Timestamp'].iloc[0].date())+ " "+ Entry_time)]
ATM_calls = ATM_calls[ATM_calls['Timestamp'] < pd.to_datetime(str(ATM_calls['Timestamp'].iloc[0].date())+ " "+ "14:59:59")]
ATM_calls.reset_index(drop=True, inplace=True)
ATM_calls
```

```python
ATM_puts = puts[puts['Ticker'].apply(lambda a : a.strip()[-7:] == ATM_put_strike)]
ATM_puts
```

```python
Entry_time = "09:30:00"
ATM_puts = ATM_puts[ATM_puts['Timestamp'] >= pd.to_datetime(str(ATM_puts['Timestamp'].iloc[0].date())+ " "+ Entry_time)]
ATM_puts = ATM_puts[ATM_puts['Timestamp'] < pd.to_datetime(str(ATM_puts['Timestamp'].iloc[0].date())+ " "+ "15:00:00")]
ATM_puts.reset_index(drop=True, inplace=True)
ATM_puts
```

```python
ohlc = {
        'Ticker':'first',
        'Open': 'first',
        'High': 'max',
        'Low':'min',
        'Close':'last',
        'Volume':'sum'
        }
    
for key in ATM_calls:
            
    ATM_calls.set_index('Timestamp', inplace=True)
    ATM_calls = ATM_calls.resample('5min').apply(ohlc)
    ATM_calls.dropna(inplace=True)
    ATM_calls.reset_index(inplace=True)
ATM_calls
```

```python
ohlc = {
        'Ticker':'first',
        'Open': 'first',
        'High': 'max',
        'Low':'min',
        'Close':'last',
        'Volume':'sum'
        }
    
for key in futures:
            
    futures.set_index('Timestamp', inplace=True)
    futures = futures.resample('5min').apply(ohlc)
    futures.dropna(inplace=True)
    futures.reset_index(inplace=True)
futures
```

```python
Entry_time = "09:30:00"
futures = futures[futures['Timestamp'] >= pd.to_datetime(str(futures['Timestamp'].iloc[0].date())+ " "+ Entry_time)]
futures = futures[futures['Timestamp'] < pd.to_datetime(str(futures['Timestamp'].iloc[0].date())+ " "+ "15:00:00")]
futures.reset_index(drop=True, inplace=True)
futures
```

```python
indicator_bb = BollingerBands(close=futures['Close'], window=20, window_dev=1.5)
#Bollinger Bands features
#x['bb_bbm'] = indicator_bb.bollinger_mavg()
futures['bb_bbh'] = indicator_bb.bollinger_hband()
futures['bb_bbl'] = indicator_bb.bollinger_lband()
futures['bb_bbh'] = futures['bb_bbh'].apply(lambda a: round(a,2))
futures['bb_bbl'] = futures['bb_bbl'].apply(lambda a: round(a,2))
#  Bollinger Band high indicator
#x['bb_bbhi'] = indicator_bb.bollinger_hband_indicator()
# Bollinger Band low indicator
#x['bb_bbli'] = indicator_bb.bollinger_lband_indicator()
futures
```

```python
futures['14-high'] = futures['High'].rolling(14).max()
futures['14-low'] = futures['Low'].rolling(14).min()
futures['%K'] = (futures['Close'] - futures['14-low'])*100/(futures['14-high'] - futures['14-low'])
futures['%D'] = futures['%K'].rolling(3).mean()
futures['%K'] = futures['%K'].apply(lambda a: round(a,2))
futures['%D'] = futures['%D'].apply(lambda a: round(a,2))
futures.drop('14-low' , axis = 1 , inplace = True)
futures.drop('14-high' , axis = 1 , inplace = True)
 
# pivot points calculations       
#futures['PP'] = (futures['High'] + futures['Low'] + futures['Close']) / 3
#futures['R1'] = 2 * futures['PP'] - futures['Low']
#futures['S1'] = 2 * futures['PP'] - futures['High']
#futures['R2'] = futures['PP'] + (futures['High'] - futures['Low'])
#futures['S2'] = futures['PP'] - (futures['High'] - futures['Low'])
#futures['R3'] = futures['PP'] + 2 * (futures['High'] - futures['Low'])
#futures['S3'] = futures['PP'] - 2 * (futures['High'] - futures['Low'])
#x.dropna(axis = 0 , inplace = True)
futures

```

```python
futures.dropna(inplace = True)
futures.reset_index(inplace = True)
futures
```

```python
#plotting the indicators on the chart
futures.plot(x="Timestamp", y=['Close','bb_bbh' , 'bb_bbl'])
futures[['%K', '%D']].plot()
```

```python
#Entry logic
futures['PE_Buy'] = 0
futures['CE_Buy'] = 0


for i in range(len(futures)):

    if futures['Close'][i] > (futures['bb_bbh'][i]+20) and futures['%K'][i] > 80 and futures['%D'][i] > 80 and futures['%K'][i] > futures['%D'][i]: #and x['Close'][i] >= x['R1'][i] or x['R2'][i] or x['R3'][i]:
        futures['PE_Buy'][i] = 1
        
    elif futures['Close'][i] < (futures['bb_bbl'][i]-20) and futures['%K'][i] < 20 and futures['%D'][i] < 20 and futures['%K'][i] < futures['%D'][i]: #and x['Close'][i] <= x['S1'][i] or x['S2'][i] or x['S3'][i]:
        futures['CE_Buy'][i] = 1
    
    else:
        pass
futures = futures[(futures['PE_Buy'] == 1) | (futures['CE_Buy'] == 1)]
futures.sort_values(by='Timestamp', inplace=True)
futures.reset_index(drop=True, inplace=True)
futures
```

```python
tradelog = pd.DataFrame(columns = ['Ticker', 'Entry Time', 'Entry Price', 'Stop Loss Exit', 'Exit Time', 'Exit Price'])
tradelog
```

```python
#STOPLOSS TARGET LOGIC
for k in range(0,len(futures)):
    if futures['PE_Buy'][k] == 1:
        Entry_PE = ATM_puts[ATM_puts['Timestamp'].apply(lambda a: a == futures['Timestamp'].iloc[k])]
        Ticker_PE = Entry_PE['Ticker'].iloc[0]
        Entry_time_PE = Entry_PE['Timestamp'].iloc[0]
        Entry_price_PE = Entry_PE['Open'].iloc[0]
        Target_price_PE = (Entry_price_PE)*((10+100)/100)
        Target_Exit_PE = ATM_puts[(ATM_puts['Open'] >= Target_price_PE) & (ATM_puts['Timestamp'] > Entry_time_PE)]     
        Stop_loss_price_PE = (Entry_price_PE)*((100-10)/100)
        Stop_loss_Exit_PE = ATM_puts[(ATM_puts['Open'] <= Stop_loss_price_PE) & (ATM_puts['Timestamp'] > Entry_time_PE)]
        print(Target_Exit_PE)
        print(Stop_loss_Exit_PE)
        
        if Stop_loss_Exit_PE.empty:
            stop_loss_exit_PE = 0
            futures['Timestamp'] = futures['Timestamp']+datetime.timedelta(minutes = 5)
            PE = ATM_puts[ATM_puts['Timestamp'].apply(lambda a: a == futures['Timestamp'].iloc[k])]
            Exit_time_PE = PE['Timestamp'].iloc[k]
            Exit_price_PE = PE['Open'].iloc[k]
        else:
            stop_loss_exit_PE = 1
            Exit_time_PE = Stop_loss_Exit_PE['Timestamp'].iloc[k]
            Exit_price_PE = Stop_loss_Exit_PE['Open'].iloc[k]
        tradelog.loc[len(tradelog)] = [Ticker_PE, Entry_time_PE, Entry_price_PE, stop_loss_exit_PE, Exit_time_PE, Exit_price_PE]
    else:
        print("NO PE TRADES")
    if futures['CE_Buy'][k] == 1:
        Entry_CE = ATM_calls[ATM_calls['Timestamp'].apply(lambda a: a == futures['Timestamp'][k])]
        Ticker_CE = Entry_CE['Ticker'].iloc[0]
        Entry_time_CE = Entry_CE['Timestamp'].iloc[0]
        Entry_price_CE = Entry_CE['Open'].iloc[0]
        Target_price_CE = (Entry_price_CE)*((10+100)/100)
        Target_Exit_CE = ATM_calls[(ATM_calls['Open'] >= Target_price_CE) & (ATM_calls['Timestamp'] > Entry_time_CE)] 
        Stop_loss_price_CE = (Entry_price_CE)*((100-10)/100)
        Stop_loss_Exit_CE = ATM_calls[(ATM_calls['Open'] <= Stop_loss_price_CE) & (ATM_calls['Timestamp'] > Entry_time_CE)]
        print(Target_Exit_CE)
        print(Stop_loss_Exit_CE)
        
        if Stop_loss_Exit_CE.empty:
            stop_loss_exit_CE = 0
            futures['Timestamp'] = futures['Timestamp']+datetime.timedelta(minutes = 5)
            CE = ATM_calls[ATM_calls['Timestamp'].apply(lambda a: a == futures['Timestamp'].iloc[k])]
            Exit_time_CE = CE['Timestamp'].iloc[k]
            Exit_price_CE = CE['Open'].iloc[k]
        else:
            stop_loss_exit_CE = 1
            Exit_time_CE = Stop_loss_Exit_CE['Timestamp'].iloc[k]
            Exit_price_CE = Stop_loss_Exit_CE['Open'].iloc[k]
        tradelog.loc[len(tradelog)] = [Ticker_CE, Entry_time_CE, Entry_price_CE, stop_loss_exit_CE, Exit_time_CE, Exit_price_CE]
    else:
        print("NO CE TRADES")
        
    
    
```

```python
tradelog
```

```python
tradelog['PnL'] = 25*(tradelog['Exit Price'] - tradelog['Entry Price'])
tradelog['Cumulative_PnL'] = tradelog['PnL'].cumsum()
tradelog
```

```python
# Adding up the pnl of every trade and plotting it
tradelog['Cumulative_PnL'].plot(title = 'Profit & Loss (PnL)', xlabel = 'Trade Number', ylabel = 'Amount', grid = True)
```
