```python
#Importing all the libs
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import datetime
import glob
import talib as ta
pd.options.mode.chained_assignment = None  
```

```python
folder = glob.glob("C:\\Users\\vaibh\\Desktop\\file(01)")
```

```python
for file in folder:
    files = glob.glob(file + '/*.csv')
files
```

```python
xx = pd.read_csv(files[0])
xx.insert(0,'Timestamp',xx['Date'] + " " + xx['Time'])
xx['Timestamp'] = pd.to_datetime(xx['Timestamp'] , format = '%d/%m/%Y %H:%M:%S')
xx.drop('Date',axis = 1 , inplace = True)
xx.drop('Time' , axis = 1 , inplace = True)
xx
```

```python
ticker = 'BANKNIFTY'
data = xx[xx['Ticker'].apply(lambda a: a[:len(ticker)] == ticker)]
data
```

```python
futures = data[data['Ticker'].apply(lambda a: a.strip()[-2:] == '-I')]
futures
```

```python
ohlc = {
        'Ticker':'first',
        'Open': 'first',
        'High': 'max',
        'Low':'min',
        'Close':'last',
        'Volume':'sum'
        }
    
for key in futures:
            
    futures.set_index('Timestamp', inplace=True)
    futures = futures.resample('5min').apply(ohlc)
    futures.dropna(inplace=True)
    futures.reset_index(inplace=True)
futures
```

```python
futures["SMA10"] = ta.SMA(futures["Close"], timeperiod=10)
futures.dropna(inplace=True)
futures.reset_index(drop ='index',inplace=True)
futures
```

```python
futures.plot(x="Timestamp", y=['Close','SMA10'])
```

```python
data = futures
data
```

```python
#Long position AND Short position
futures['Long'] = 0
futures['Short'] = 0
for i in range(len(futures)-1):
    if futures['Close'].iloc[i]>futures['SMA10'].iloc[i] and futures['Close'].iloc[i+1]>futures['Close'].iloc[i]:
        futures['Long'].iloc[i] = 1
        futures['Short'].iloc[i] = 0 
    elif futures['Close'].iloc[i]<futures['SMA10'].iloc[i] and futures['Close'].iloc[i+1]<futures['Close'].iloc[i]:
        futures['Long'].iloc[i] = 0
        futures['Short'].iloc[i] = 1
    else:
        pass
futures = futures[(futures['Long'] == 1) | (futures['Short'] == 1)]
futures.sort_values(by='Timestamp', inplace=True)
futures.reset_index(drop=True, inplace=True)
futures
```

```python
data.drop(['Long','Short'],axis = 1,inplace = True)
data
```

```python
#target and stoploss
futures['Buy_long'] = 0
futures['Sell_long'] = 0
futures['Short_sell'] = 0
futures['Buyback_short'] = 0
for j in range(len(futures)):
    if futures['Long'][j] == 1 and futures['Short'][j] == 0:
        futures['Buy_long'][j] = 1 
        if futures['Close'][j] < futures['Close'][j+1]:
            futures['Sell_long'][j] = 1
            stop_loss_time_long = futures['Timestamp'].iloc[j]-datetime.timedelta(minutes = 5)
            stop_loss_price_long = futures[futures['Timestamp'].apply(lambda a: a==stop_loss_time_long)]['Close']
        elif futsures['Close'][j] > futures['Close'][j+1]:
            futures['Sell_long'][j] = 1
            target_time_long = futures['Timestamp'].iloc[j]+datetime.timedelta(minutes = 5)
            target_price_long = futures[futures['Timestamp'].apply(lambda a: a ==target_time_long)]['Close']
    elif futures['Short'][j] == 1 and futures['Long'][j] == 0:
        futures['Short_sell'][j] = 1 
        if futures['Close'][j] > futures['Close'][j+1]:
            futures['Buyback_short'][j] = 1
            stop_loss_time_long = futures['Timestamp'].iloc[j]-datetime.timedelta(minutes = 5)
            stop_loss_price_long = futures[futures['Timestamp'].apply(lambda a: a==stop_loss_time_long)]['Close']
        elif futures['Close'][j] < futures['Close'][j+1]:
            futures['Buyback_short'][j] = 1
            target_time_long = futures['Timestamp'].iloc[j]+datetime.timedelta(minutes = 5)
            target_price_long = futures[futures['Timestamp'].apply(lambda a: a == target_time_long)]['Close']     
    else:
        pass
```
